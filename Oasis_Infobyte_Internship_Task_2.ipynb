Checking the Correlation between the feature of dataset

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 6))
ax=sns.heatmap(df.corr(),annot=True,linewidths=3)
ax.tick_params(size=10,labelsize=10,labelcolor='w')

plt.show()
```

```python
plt.title("Unemployment in India")
sns.histplot(x='Estimated Employed',hue='Region',data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate',hue='Region',data=df)
plt.show()
```

```python
Unemployment=df[['States','Region','Estimated Unemployment Rate']]
figure=px.sunburst(Unemployment,path=['Region','States'],values='Estimated Unemployment Rate',width=500,height=500,color_continuous_scale='RdY1Gn',title='Unemployment Rate in India')
figure.show()
```
